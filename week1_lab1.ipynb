```python
scores = [72, 85, 90, 68, 88]
```

1. Printing all scores

```python
print("Scores: ", scores)
```

2. Print average score


```python
average = sum(scores) / len(scores)
print("Average: ", average)
```

Problem 3

```python
def pass_fail(score):
  if score >= 60:
    return "Pass"
  else:
    return "Fail"

scores = [45 , 67, 82, 59]

for score in scores:
  print("Score: ",score,"->",pass_fail(score))
```
